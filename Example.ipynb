```python
#Write a program that takes two integers, computes their sum, difference, product, and division, checks if they’re even/odd, and converts one to a float.
a = int(input("Enter first integer: "))
b = int(input("Enter second integer: "))

sum_result = a + b
difference = a - b
product = a * b

if b != 0:
    division = a / b
else:
    division = "Undefined (cannot divide by zero)"

def check_even_odd(num):
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"

a_float = float(a)

print("\nResults:")
print("Sum:", sum_result)
print("Difference:", difference)
print("Product:", product)
print("Division:", division)

print("\nEven/Odd Check:")
print(f"{a} is", check_even_odd(a))
print(f"{b} is", check_even_odd(b))

print("\nFloat Conversion:")
print(f"{a} as float is:", a_float)
```

```python
sentence = input("Enter a sentence: ")

vowels = "aeiouAEIOU"
vowel_count = 0
consonant_count = 0

for ch in sentence:
    if ch.isalpha():
        if ch in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

reversed_sentence = sentence[::-1]

modified_sentence = sentence.replace(" ", "_")

capitalized_sentence = sentence.title()

print("\nResults:")
print("Vowel count:", vowel_count)
print("Consonant count:", consonant_count)
print("Reversed sentence:", reversed_sentence)
print("Spaces replaced:", modified_sentence)
print("Capitalized sentence:", capitalized_sentence)
```

```python
t = (10, "hello", 3.5, "world", 7, True)

numeric_values = tuple(x for x in t if isinstance(x, (int, float)) and not isinstance(x, bool))

print("Original tuple:", t)
print("Numeric values:", numeric_values)

try:
    t[0] = 100   
except TypeError as e:
    print("\nError while modifying tuple:", e)

t2 = ("Python", 99)
concatenated = t + t2

print("\nSecond tuple:", t2)
print("Concatenated tuple:", concatenated)
```

```python
student_marks = {
    "Piyush": 85,
    "Rahul": 78,
    "Ankit": 90
}

print("Initial Dictionary:", student_marks)

student_marks["Neha"] = 88
print("\nAfter Adding Neha:", student_marks)
student_marks["Rahul"] = 82
print("\nAfter Updating Rahul:", student_marks)

del student_marks["Ankit"]
print("\nAfter Deleting Ankit:", student_marks)

print("\nKeys:", student_marks.keys())
print("Values:", student_marks.values())
print("Items:", student_marks.items())
```

```python
strings = ["level", "hello world", "python", "madam", "data science"]

print("Original list:", strings)

sorted_strings = sorted(strings, key=lambda x: len(x))
print("\nSorted by length:", sorted_strings)

palindromes = [s for s in strings if s.replace(" ", "").lower() == s.replace(" ", "").lower()[::-1]]
print("\nPalindromes:", palindromes)

modified_strings = [s.replace(" ", "-") for s in strings]
print("\nSpaces replaced with hyphens:", modified_strings)
```
